```python
#NORMAL DISTRIBUTION
import numpy as np 
import matplotlib.pyplot as plt

values= np.random.normal(300,25, 10000)
print(values.shape)
plt.hist(values,100,density=True)
plt.show()
```

```python
#UNIFORM DISTRIBUTION
s = np.random.uniform(-8,10,1000)
count, bins, ignored = plt.hist(s, 60,density=True)
plt.show()
```

```python
#POISSON DISTRIBUTION
pois = np.random.poisson(7.9, 12000)
plt.hist(pois, 10,density=True)
plt.show()
```

```python
#BINOMIAL DISTRIBUTION
n=100
p=0.1667
binom = np.random.binomial(n, p, 10000)
plt.hist(binom,20,density=True)
plt.show()
```

```python
#EXPONENTIAL DISTRIBUTION
target = 250
beta = 1.0/target
Y = np.random.exponential(beta, 5000)
plt.hist(Y, bins=200,lw=0,alpha=.8)
plt.plot([0,max(Y)],[target,target],'r--')
plt.ylim(0,target*1.1)
plt.show()
```

```python
#RANDOM DISTRIBUTION
ran=np.random.choice(range(8000,10000),100)
ran
plt.hist(ran,5)
plt.show()
```
